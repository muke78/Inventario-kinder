```python
# Cargar el archivo CSV
try:
    xlxsCargado = pd.read_csv(input_file, encoding='unicode_escape')
except Exception as e:
    print(f"Error al cargar el archivo: {e}")
    exit()
```

```python
# Validar columnas necesarias
columnas_necesarias = [
    "NIC", "Numero inventario", "Nombre Articulo", "Marca", 
    "Modelo", "Valor", "Fecha de Asignacion"
]
```

```python
for columna in columnas_necesarias:
    if columna not in xlxsCargado.columns:
        raise ValueError(f"La columna '{columna}' no está presente en el archivo CSV.")
```

```python
# Asegurar que la columna "Numero inventario" sea de tipo string
xlxsCargado["Numero inventario"] = xlxsCargado["Numero inventario"].astype(str)

# Filtrar por Número de inventario
inventario_filtrar = [str(num) for num in inventario_filtrar]  # Convertir filtro a cadenas
df_filtrado = xlxsCargado[xlxsCargado["Numero inventario"].isin(inventario_filtrar)]
```

```python
def estado_fecha(asignacion):
    try:
        # Convertir a formato de fecha
        fecha = pd.to_datetime(asignacion, format="%d/%m/%Y", errors='coerce')
        if pd.isnull(fecha):
            return "Fecha inválida"
        return "Aprobado para dar de baja" if fecha.year <= 2007 else "No aprobado"
    except:
        return "Fecha inválida"


df_filtrado = df_filtrado.copy()
df_filtrado["Estado Baja"] = df_filtrado["Fecha de Asignacion"].apply(estado_fecha)
# Contar las repeticiones de cada artículo
conteo_articulos = df_filtrado["Nombre Articulo"].value_counts()

# Sumar todos los valores del conteo
suma_total = conteo_articulos.sum()

# Mostrar el conteo de artículos
conteo_articulos
suma_total
```

```python
# Graficar los resultados como un gráfico de barras
plt.figure(figsize=(10, 8))
ax = conteo_articulos.plot(kind='bar', color='orange')

# Añadir títulos y etiquetas
plt.title("Conteo de Artículos para Baja", fontsize=14)
plt.xlabel("Nombre Artículo", fontsize=12)
plt.ylabel("Cantidad", fontsize=12)

# Rotar las etiquetas del eje X para mejorar la legibilidad
plt.xticks(rotation=45, ha='right')

# Agregar etiquetas sobre las barras
for i, v in enumerate(conteo_articulos):
    ax.text(i, v + 0.2, str(v), ha='center', va='bottom', fontsize=10)

# Agregar la suma total al gráfico
ax.text(len(conteo_articulos) - 1, suma_total + 10, f"Suma total: {suma_total}", 
        ha='center', va='bottom', fontsize=12, color='red', fontweight='bold')

# Ajustar el diseño para que no se corte texto
plt.tight_layout()  
plt.savefig('Grafico_conteo_articulos_para_baja.png')
plt.show()
```

```python
# Convertir la columna "Valor" a formato de moneda
def formato_moneda(valor):
    try:
        return f"${valor:,.2f}"  # Formato con comas y dos decimales
    except:
        return "Valor inválido"

df_filtrado["Valor"] = df_filtrado["Valor"].apply(formato_moneda)
```

```python
# Seleccionar las columnas finales
columnas_finales = columnas_necesarias + ["Estado Baja"]
df_limpio = df_filtrado[columnas_finales]
df_limpio
```
